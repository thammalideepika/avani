```python
import pandas as pd
df = pd.read_excel('EVEN.xlsx', sheet_name='Sheet1')

average_cost_by_destination = df.groupby('Destination')['Total_Cost'].mean().reset_index()
print(average_cost_by_destination)
```

```python
payment_method_distribution = df['Payment_Method'].value_counts().reset_index()
payment_method_distribution.columns = ['Payment_Method', 'Count']
print(payment_method_distribution)
```

```python
unique_destinations = df['Destination'].nunique()
print(f"Number of unique destinations: {unique_destinations}")
```

```python
# Check for missing values in important columns
missing_values = df[['Travel_Package', 'Payment_Method', 'Travel_Agent']].isnull().sum()
print(missing_values)
```
